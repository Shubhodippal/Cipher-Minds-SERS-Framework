```python
kuet_folder = glob.glob(kuet+"\*")
kuet_folder
```

```python
kuet_file_name = []
kuet_file_path = []
for i in range (0, len((kuet_folder))):
    kuet_file_name.append(os.listdir(kuet_folder[i]))
    kuet_file_path.append(glob.glob(kuet_folder[i]+"\*"))

kuet_file_name = list(itertools.chain(*kuet_file_name))
kuet_file_name
```

```python
kuet_file_path = list(itertools.chain(*kuet_file_path))
kuet_file_path
```

```python
kuet_emotion = []
for i in range(0, len(kuet_file_path)):
    em = (kuet_file_path[i].split('\\')[7][0])
    if em == 'A':
        kuet_emotion.append('1')
    elif em == 'D':
        kuet_emotion.append('2')
    elif em == 'H':
        kuet_emotion.append('4')
    elif em == 'N':
        kuet_emotion.append('7')
    elif em == 'S':
        kuet_emotion.append('5')
    else:
        break
kuet_emotion
```

```python
kuet_df = pd.DataFrame(kuet_file_path, columns=['File Path'])
kuet_df['File Name'] = kuet_file_name
kuet_df['Emotion'] = kuet_emotion
kuet_df
```

```python
b_ser_folder = (glob.glob(b_ser+"\*"))
b_ser_folder
```

```python
b_ser_file_path = []
b_ser_file_name = []
for i in range(0, len(b_ser_folder)):
    b_ser_file_path.append(glob.glob(b_ser_folder[i]+"\*"))
    b_ser_file_name.append(os.listdir(b_ser_folder[i]))

b_ser_file_path = list(itertools.chain(*b_ser_file_path))
b_ser_file_name = list(itertools.chain(*b_ser_file_name))
b_ser_file_path
```

```python
b_ser_file_name
```

```python
b_ser_emotion = []
for i in range(0, len(b_ser_file_name)):
    b_ser_em = b_ser_file_name[i].split("-")[2][1]
    #print(b_ser_em)
    if b_ser_em == '1':#happy
        b_ser_emotion.append('4')
    elif b_ser_em == '2':#sad
        b_ser_emotion.append('5')
    elif b_ser_em == '3': # anger
        b_ser_emotion.append('1')
    elif b_ser_em == '4': # surprised
        b_ser_emotion.append('6')
    elif b_ser_em == '5': # neutral
        b_ser_emotion.append('7')
b_ser_emotion
```

```python
b_ser_df = pd.DataFrame(b_ser_file_path, columns=['File Path'])
b_ser_df['File Name'] = b_ser_file_name
b_ser_df['Emotion'] = b_ser_emotion
b_ser_df
```

```python
subesco_folder = (glob.glob(subesco+"/*"))
subesco_folder
```

```python
subesco_file_names = os.listdir(subesco)
subesco_file_names
```

```python
subesco_emotion = []
for i in range(0, len(subesco_file_names)):
    subesco_em = (subesco_file_names[i].split('_')[5].split('.')[0])
    #print(subesco_em)
    if subesco_em == 'ANGRY':
        subesco_emotion.append('1')
    elif subesco_em == 'DISGUST':
        subesco_emotion.append('2')
    elif subesco_em == 'FEAR':
        subesco_emotion.append('3')
    elif subesco_em == 'HAPPY':
        subesco_emotion.append('4')
    elif subesco_em == 'NEUTRAL':
        subesco_emotion.append('7')
    elif subesco_em == 'SAD':
        subesco_emotion.append('5')
    elif subesco_em == 'SURPRISE':
        subesco_emotion.append('6')
subesco_emotion
```

```python
subesco_df = pd.DataFrame(subesco_folder, columns=['File Path'])
subesco_df['File Name'] = subesco_file_names
subesco_df['Emotion'] = subesco_emotion
subesco_df
```

```python
df = pd.concat([Banspemo_df, kuet_df, b_ser_df, subesco_df], axis=0, ignore_index=True)
df
```
